```python
# Convert "sex" column to numeric
df_mandena_filtered.loc[:, "sex"] = df_mandena_filtered["sex"].map({"female": 0, "male": 1})
# Scale "mass", "asv_degree", "elevation.obs" and "host_degree" columns
scaler = StandardScaler()
df_mandena_filtered.loc[:, "mass"] = scaler.fit_transform(df_mandena_filtered[["mass"]])
df_mandena_filtered.loc[:, "asv_degree"] = scaler.fit_transform(df_mandena_filtered[["asv_degree"]])
df_mandena_filtered.loc[:, "elevation.obs"] = scaler.fit_transform(df_mandena_filtered[["elevation.obs"]])
df_mandena_filtered.loc[:, "host_degree"] = scaler.fit_transform(df_mandena_filtered[["host_degree"]])

# Get dummies for "grid" column
df_mandena_filtered = pd.get_dummies(df_mandena_filtered, columns=["grid"], dtype="int", drop_first=True)
df_mandena_filtered.head()
```

## Choosing a model

```python
X = df_mandena_filtered.drop(["Link", "asv_ID"], axis=1)
y = df_mandena_filtered["Link"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier()
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

mcc_scorer = make_scorer(matthews_corrcoef)

# Dictionary to hold the MCC results
mcc_results = {model_name: [] for model_name in models.keys()}

# Perform cross-validation and calculate MCC scores
for model_name, model in models.items():
    cv_results = cross_val_score(model, X_train, y_train, cv=kf, scoring=mcc_scorer)
    mcc_results[model_name] = cv_results
```

```python
# create confusion matrix after cross validation
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, annot_kws={'size': 18}, fmt='d', cmap='Oranges', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted', weight='bold', fontsize=14)
plt.ylabel('Actual', weight='bold', fontsize=14)
plt.title('Confusion Matrix - Logistic Regression', weight='bold', fontsize=20)
precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])
print("Precision:", precision)
recall = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print("Recall:", recall)
plt.savefig('ConfusionMatrixLR.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# create confusion matrix after cross validation
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, annot_kws={'size': 18}, fmt='d', cmap='Oranges', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted' , weight='bold', fontsize=14)
plt.ylabel('Actual', weight='bold', fontsize=14)
plt.title('Confusion Matrix - KNN', weight='bold', fontsize=20)
plt.savefig('ConfusionMatrixKNN.png', bbox_inches='tight', dpi=300)
# precision
precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])
print("Precision:", precision)
# recall
recall = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print("Recall:", recall)
plt.show()
```

```python
# Convert results to DataFrame for plotting
results_df = pd.DataFrame(mcc_results)

# Melt the DataFrame for Seaborn
results_melted = results_df.melt(var_name='Model', value_name='MCC')

# Plotting the MCC scores
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8), dpi=300)
boxplot = sns.boxplot(x='Model', y='MCC', data=results_melted, palette="Set2")
boxplot.set_title('Comparison of MCC Scores Across Models', fontsize=20, weight='bold')
boxplot.set_xlabel('Model', fontsize=15, weight='bold')
boxplot.set_ylabel('MCC Score', fontsize=15, weight='bold')

boxplot.set_xticklabels(boxplot.get_xticklabels(), fontsize=22, weight='bold', color='black')

plt.savefig('mcc_comparison_boxplot.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def custom_beeswarm_plot(shap_values, features, feature_names, max_display=11, plot_range=(-2, 2)):
    shap_values = shap_values.values
    feature_names = np.array(feature_names)

    # Compute the mean absolute SHAP values
    mean_abs_shap = np.abs(shap_values).mean(0)

    # Get the indices of features sorted by their mean absolute SHAP values
    feature_order = np.argsort(mean_abs_shap)[::-1]

    # Identify the indices of the important features
    important_features = ['asv_degree', 'host_degree']
    important_indices = [np.where(feature_names == feature)[0][0] for feature in important_features]

    # Ensure important features are included
    included_indices = list(feature_order[:max_display])

    # Add important features if not already included
    for idx in important_indices:
        if idx not in included_indices:
            if len(included_indices) < max_display:
                included_indices.append(idx)
            else:
                # Replace the least important feature with the important one
                included_indices[-1] = idx

    # Sort the final indices by their mean absolute SHAP values to maintain importance
    final_indices = sorted(included_indices, key=lambda x: mean_abs_shap[x], reverse=True)

    plt.figure(figsize=(10, 8))

    # Plot the features in reverse order
    for i, feature in enumerate(reversed(final_indices)):
        values = shap_values[:, feature]
        feature_values = features[:, feature]
        mask = (values >= plot_range[0]) & (values <= plot_range[1])

        plt.scatter(values[mask], [i] * len(values[mask]),
                    c=feature_values[mask], cmap='coolwarm', s=5)

    plt.yticks(range(len(final_indices)),
               [feature_names[i] for i in reversed(final_indices)])
    plt.xticks(weight='bold')
    plt.xlabel('SHAP value (impact on model output)')
    plt.ylabel('Feature')
    plt.title(f'Zoom in to range: ({plot_range[0]} to {plot_range[1]})', weight='bold', fontsize=20)
    plt.colorbar(label='Feature value')
    plt.grid(True, alpha=0.3)
    plt.savefig('ZoomInSHAP.png', dpi=300, bbox_inches='tight')
    plt.show()

# Example usage:
custom_beeswarm_plot(shap_values, X_test_resampled.values, X_test_resampled.columns, max_display=10, plot_range=(-2, 2))
```
